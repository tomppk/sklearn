## 0. An end-to-end Scikit-Learn workflow

```python
# 1. Get the data ready
import pandas as pd
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
import numpy as np
```

```python
# Create X (features matrix). Every column except "target" column which we are trying to predict
x = heart_disease.drop("target", axis=1)

# Create y (labels)
y = heart_disease["target"]
```

```python
# 2. Choose the right model and hyperparameters
# Our problem is classification, trying to predict whether patient has heart disease or not
# Random Forest is a model used for classification problems
# clf is short for classification
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# Keep the default hyperparameters. See what default params are
clf.get_params()
```

```python
# 3. Fit the model to the training data. Split the data into training data and test data
# test_size=0.2 means 20% of data will be used for testing, 80% for training
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# Tells Random Forest model to find patterns in training data
# Fits the model to the training data
clf.fit(x_train, y_train);
```

```python
# Show training data
x_train
```

```python
# Show test data
x_test
```

```python
# As the model has been trained. Make predictions in the test data
y_preds = clf.predict(x_test)
y_preds
```

```python
y_test
```

```python
# 4. Evaluate the model on the training data and test data
clf.score(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
# Evaluating the model with different metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Print classification report that compares test labels ie. "target" to prediction labels
print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
# 5. Improve a model
# Try different amount of n_estimators
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(x_train, y_train)
    print(f"Model accuracy on test set: {clf.score(x_test, y_test) * 100:.2f}%")
    print("")
```

```python
# 6. Save a model and load it
import pickle

pickle.dump(clf, open("random_forest_model_1.pkl", "wb"))
```

```python
# Import model
loaded_model = pickle.load(open("random_forest_model_1.pkl", "rb"))
loaded_model.score(x_test, y_test)
```
